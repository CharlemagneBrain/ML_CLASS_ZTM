## Matplotlib example workflow


```python
# 0. import matplotlib and get ready
%matplotlib inline
import matplotlib.pyplot as plt 

#1. Prepare data
x=[1,2,3,4,5]
y=[11,12,13,14,15]

# 2.Setup plot

fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot
ax.plot(x,y);

# 4. Personnaliser nos trac√©s
ax.set(
        title="Simple Trac√©",
        xlabel="x-axis",
        ylabel="y-axis",      
      )

# 5. Save the plot

fig.savefig("images/simple-plot.png")
```

```python
# Make a plot from dictionnary

nut_butter_prices = {
    "Almond Butter" : 10,
    "Peanut Butter" : 8,
    "Cashew Butter" : 12
}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(
        title="Charlie's Nut Butter Store",
        ylabel="Prices",
        xlabel="Butters"
      );
```

```python
# Barres horizontales (Transformer nos donn√©es en listes)

fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Deux options de subplots

```python
# 1√®re option
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(
                                                nrows = 2,
                                                ncols=2,
                                                figsize=(10,10)
                                                );
# Trac√© pour chaque axe

ax1.plot(x, x/2);
ax2.hist(nut_butter_prices);
ax3.scatter(np.random.random(10), np.random.random(10));
ax4.bar(nut_butter_prices.keys(), nut_butter_prices.values());
```

```python
# 2i√®me option
fig, ax = plt.subplots(
                                                nrows = 2,
                                                ncols=2,
                                                figsize=(10,10)
                                                );

ax[0,0].hist(x);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].plot(x, x/2);
ax[1,1].bar(nut_butter_prices.keys(), nut_butter_prices.values());
```

## Lequel utiliser ?

* Pour effectuer un trac√© rapide, utiliser la m√©thode pyplot
* Pour avoir trac√© plus avanc√©, utiliser la m√©thode matplotlib OO

```python
heart_disease.head()
```

```python
over_50 = heart_disease[ heart_disease["age"] > 50 ]
over_50.head()
```

```python
len(over_50)
```

```python
# Pyplot method => tra√ßage directe

over_50.plot(kind="scatter", x="age", y="chol", c="target", figsize=(10,6));
```

```python
# OO Method mixed with pyplot

fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(
                kind="scatter",
                x="age",
                y="chol",
                c="target",
                ax=ax
            );

# ax.set_xlim([45,100])
```

```python
# OO Method from scratch using OOP
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data

scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"]);

# Customize the plot

ax.set(title="Maladie Cardiaque & Taux de Cholest√©rol",
      xlabel="Age", ylabel="Cholest√©rol");

# Ajouter une l√©gende
ax.legend(*scatter.legend_elements(), title="Target");

# Ajouter une ligne horizontale d√©signant le taux moyen de cholest√©rol

ax.axhline(over_50["chol"].mean(), linestyle='--');
```

```python
over_50.head()
```

# Graphique repr√©sentant le taux de cholest√©rol, le rythme cardiaque en fonction de l'√¢ge des patients

```python
# Subplot of chol, thalach and age
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(15,10), sharex=True)

# Associer des donn√©es √† l'axe0

scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"]);

# Personalisation de la figure

ax0.set(title="Maladie Cardiaque et Taux de Cholest√©rol", ylabel="Cholesth√©rol");

# Ajoutons une l√©gende
ax0.legend(*scatter.legend_elements(), title="Target");

# Ajoutons la barre horizontale d√©signant le taux moyen de cholesth√©rol

ax0.axhline(y=over_50["chol"].mean(), linestyle="--");

#-------------------------------------------------------------------------------------------------------#

# Associer des donn√©es √† l'axe1

scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"]);

# Personalisation de la figure

ax1.set(title="Analyse de la Fr√©quence cardiaque maximale", xlabel="√Çge", ylabel="Rythme Cardiaque");

# Ajoutons une l√©gende 
ax1.legend(*scatter.legend_elements(), title="Target");

# Ajoutons la barre horizontale d√©signant le taux moyen de cholesth√©rol

ax1.axhline(y=over_50["thalach"].mean(), linestyle="--");

# Ajouter un titre √† la figure

fig.suptitle("Analyse des Maladies Cardiaques", fontsize=16, fontweight="bold");

```
